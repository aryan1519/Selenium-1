```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome(executable_path = 'chromedriver_win32/chromedriver.exe' )
```

```python
driver.get('https://www.codingninjas.in/') 
```

```python
driver.title
```

```python
# Clicking back button
driver.back()
```

```python
# Clicking forward button
driver.forward()
```

```python
# Taking screenshot
driver.save_screenshot('mySS.png')
```

```python
# Getting the html content

driver.page_source
```

```python
# Find hyperlink by its text

My_Classroom_button = driver.find_element_by_link_text('My Classroom')

My_Classroom_button.click()
```

```python
# Find hyperlink by partial text 

My_Classroom_button = driver.find_element_by_partial_link_text('My')

My_Classroom_button.click()
```

```python
# Maximize window   
driver.maximize_window()
```

```python
# Refresh
driver.refresh()
```

```python
# Getting the current url

driver.current_url
```

```python

# Close and quit
# close closes current window quit closes all windows
driver.close
driver.quit()
```
